```python
import tensorflow as tf
import keras
import matplotlib.pyplot as plt
import pandas as pd
import os
import shutil
import matplotlib.image as mpimg
import numpy as np
import keras
```

```python
class TFRecordExtractor:
    def __init__(self, tfrecord_file):
        self.tfrecord_file = os.path.abspath(tfrecord_file)
        

    def _extract_fn(self, tfrecord):
        # Extract features using the keys set during creation
        features = {"image/height":    tf.io.FixedLenFeature([], tf.int64),
        "image/width":    tf.io.FixedLenFeature([], tf.int64), 
        "image/colorspace": tf.io.FixedLenFeature([], tf.string),
        "image/channels": tf.io.FixedLenFeature([], tf.int64),
        "image/class/label": tf.io.FixedLenFeature([], tf.int64),
        "image/class/raw": tf.io.FixedLenFeature([], tf.int64),
        "image/class/source": tf.io.FixedLenFeature([], tf.int64),
        "image/class/text":tf.io.VarLenFeature(tf.string),     
        "image/format": tf.io.VarLenFeature(tf.string),
        "image/filename": tf.io.VarLenFeature(tf.string), 
        "image/id":tf.io.FixedLenFeature([], tf.int64),
        "image/encoded":  tf.io.FixedLenFeature([],tf.string)
        }

        # Extract the data record
        sample = tf.io.parse_single_example(tfrecord, features)
        image = tf.image.decode_jpeg(sample['image/encoded'],channels=3) 
        reshaped = tf.image.resize(image,[599,599],preserve_aspect_ratio=False)
        reshaped = tf.cast(reshaped,tf.uint8)
        label = sample['image/class/label']
        filename = sample['image/filename']
    
        return reshaped, label, filename        

    def extract_image(self,num): #to check the nth image in the dataset
        # Create folder to store extracted images

        # Pipeline of dataset and iterator 
        dataset = tf.data.TFRecordDataset([self.tfrecord_file])
        dataset = dataset.map(self._extract_fn)
        iterator = iter(dataset)
        for i in range(num):
                    image,label,filename = iterator.get_next()
                    #if(image_data[1].numpy()==2):
                    plt.imshow(image,interpolation='none')
                    print("Image_shape:"+str(np.shape(image)))            
        return dataset
    
    ## FUNCTIONS TO FIND OUT ABOUT THE DATASET (Features like count, size of each image,etc.)
    def create_np_data(self,dataset):# to create numpy array of the data
        iterator = iter(dataset)
        
        np_dataset = []        
        labels =[]        
        for element in dataset:
                img,label,filenames = iterator.get_next()
                
                np_dataset.append(img)
                labels.append(label)
        print("Appended")
        return np_dataset,labels
        
    def check_data(self,dataset):# to check the different sizes/ shapes present in  the data
        iterator = iter(dataset)
        k=0
        sizes = []
        for element in dataset:
            image_data = iterator.get_next()
            shapex = [np.shape(image_data[0].numpy())]
            if shapex not in sizes:
                sizes.append(shapex)
            k+=1
        
        print("Size of dataset is :"+str(k)+"\n") 
        print(sizes)
        print("\n")

    
```

```python
q = TFRecordExtractor('train-00001-of-00407')
dataset1 = q.extract_image(10)
q.check_data(dataset1) 
```

```python
x_train,y_train = q.create_np_data(dataset1)
```

```python
x_train = tf.stack(x_train,axis=0)
x_train
```

```python
def show_any_image(index):
    msg ="Type is:"
    if y_train[index]==0:
        msg+="Clear"
    elif y_train[index]==1:
        msg+="Crystal"
    elif y_train[index]==2:
        msg+="Precipitate"
    else:
        msg+="Others"
    
    plt.imshow(x_train[index])
    plt.title(msg)
```

```python
show_any_image(300)
```

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D,Input
input_tensor = Input(shape=(599, 599, 3))
# create the base pre-trained model
base_model = InceptionV3(weights='imagenet', include_top=False,input_tensor=input_tensor)
```

```python
print(base_model.summary())
```

```python
from tensorflow.keras.layers import AveragePooling2D,Flatten,Dropout
from tensorflow.keras.optimizers import Adam
headModel = base_model.output
headModel = AveragePooling2D(pool_size=(4, 4))(headModel)
headModel = Flatten(name="flatten")(headModel)
headModel = Dense(128, activation="relu")(headModel)

headModel = Dense(128, activation="relu")(headModel)
headModel = Dropout(0.5)(headModel)
headModel = Dense(4, activation="softmax")(headModel)
```

```python
model = Model(inputs=base_model.input, outputs=headModel)
```

```python
model.summary()
```

```python
for layer in base_model.layers:
	layer.trainable = False
```

```python
opt = Adam(lr=1e-4)
model.compile(loss="binary_crossentropy", optimizer=opt,
	metrics=["accuracy"])
```

```python
y_train = tf.stack(y_train)
```

```python
y_train
```

```python
model.fit(x_train,y_train,verbose=2,epochs=4)
```
